```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Bank_customer_data.csv')

print(data.info())
print(data.describe())
print(data.head())
```

```python
print(data.isnull().sum())  # Count missing values in each column
```

```python
sns.countplot(x=data["Personal Loan"])
plt.title("Loan Approval Distribution")
plt.show()

print(data["Personal Loan"].value_counts(normalize=True))  # Check class percentages
```

```python
# Histogram of Numerical Features

data.hist(figsize=(12, 8), bins=20, edgecolor="black")
plt.tight_layout()
plt.show()
```

```python
# Box Plots (Detect Outliers in Numerical Features)

plt.figure(figsize=(12, 6))
sns.boxplot(data=data.drop(columns=["Personal Loan"]))  # Exclude target variable
plt.xticks(rotation=45)
plt.title("Box Plot of Numerical Features")
plt.show()
```

```python
# Density Plots (Understand Feature Distributions)

plt.figure(figsize=(12, 6))
for col in ["Income", "CCAvg", "Mortgage"]:
    sns.kdeplot(data[col], label=col, fill=True)  # Use fill=True instead of shade=True
plt.title("Density Plots of Key Financial Features")
plt.legend()
plt.show()
```

```python
# Pair Plot (Explore Feature Relationships)

sns.pairplot(data, hue="Personal Loan", diag_kind="kde")
plt.show()
```

```python
# Bar Chart for Categorical Feature

sns.countplot(x=data["Education"], hue=data["Personal Loan"])
plt.title("Education Level vs. Loan Approval")
plt.show()
```
